```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 1: Create data
data = {
    "Month": [1,2,3,4,5],
    "Sales": [200,250,300,280,350]
}

df = pd.DataFrame(data)

# Step 2: Define input and output
X = df[['Month']]
y = df['Sales']

# Step 3: Train model
model = LinearRegression()
model.fit(X, y)

# Step 4: Predict future
future = pd.DataFrame({'Month':[6,7,8,9]})
pred = model.predict(future)

print("Future Sales:", pred)

# Step 5: Plot graph
plt.scatter(X, y)
plt.plot(future, pred)
plt.title("Sales Forecast")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.savefig("forecast.png")  # saves image
plt.show()
```
